```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read training dataset
train_df = pd.read_csv('dataset/train.csv')
```

```python
def plot_pie(df: pd.DataFrame, column_name: str):
    # Count the occurrences of each label
    label_counts = df[column_name].value_counts()
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        label_counts,
        labels=label_counts.index,
        autopct='%1.1f%%',  # Show percentage with 1 decimal
        startangle=140,
        textprops={'fontsize': 12}
    )
    
    plt.title(f'Label Distribution in {column_name} Column', fontsize=14)
    plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
    plt.show()
```

```python
plot_pie(df=train_df, column_name="label_name")
```

```python
# Read testing dataset
test_df = pd.read_csv('dataset/test.csv')
```

```python
plot_pie(df=test_df, column_name="label_name")
```

```python
# Read validation dataset
validation_df = pd.read_csv('dataset/validation.csv')
```

```python
plot_pie(df=validation_df, column_name="label_name")
```

```python
# Concatenate along rows (axis=1)
df_merged = pd.concat([train_df, test_df, validation_df], axis=0, ignore_index=True)
```

```python
plot_pie(df=df_merged, column_name="label_name")
```
